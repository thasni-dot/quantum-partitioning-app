```python
import matplotlib.pyplot as plt
```

```python
bruteforce_cuts = 5
bruteforce_time = 18.136102199554443

kl_cuts = 7
kl_time = 0.0009984970092773438

spectral_cuts = 7
spectral_time = 1.7231626510620117
```

```python
algorithms = [
    "Brute Force",
    "Kernighan-Lin",
    "Spectral"
]

cut_counts = [
    bruteforce_cuts,
    kl_cuts,
    spectral_cuts
]

runtimes = [
    bruteforce_time,
    kl_time,
    spectral_time
]

```

```python
print("Algorithm Comparison")
print("-" * 50)

for algo, cuts, runtime in zip(
    algorithms,
    cut_counts,
    runtimes
)

    print(
        f"{algo:15}  Cuts = {cuts:3d}  Runtime = {runtime:.6f} s"
    )
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    algorithms,
    cut_counts
)

plt.title("Cut Count Comparison")
plt.ylabel("Number of Cut Gates")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    algorithms,
    runtimes
)

plt.title("Runtime Comparison")
plt.ylabel("Runtime (seconds)")

plt.show()
```
